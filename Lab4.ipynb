```python
euro = pd.read_csv('Euro_2012_stats.csv')
```

```python
euro_goal = euro.sort_values('Goals', ascending = False).head(5)
euro_goal
```

```python
euro['shots_diff_on_to_off'] = euro['Shots on target'] - euro['Shots off target']
euro
```

```python
total_euro_goals = euro['Goals'].sum()
total_euro_goals
```

```python
total_euro_penalty_goals = euro['Penalty goals'].sum()
total_euro_penalty_goals_not_scored = euro['Penalties not scored'].sum()
total_euro_penalty_scored_percent_goals = total_euro_penalty_goals * 100 / (total_euro_penalty_goals + total_euro_penalty_goals_not_scored)

total_euro_penalty_scored_percent_goals
```

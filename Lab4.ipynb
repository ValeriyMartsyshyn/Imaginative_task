```python
goal_difference = euro['Goals'] - euro['Goals conceded']
goal_difference.index = euro['Team']
print(goal_difference.sort_values(ascending=False))
```

```python
shots_on_target = euro['Saves made']+euro['Goals conceded'] #Shots on target on team's gates
```

```python
penalty_scoring_chance = euro['Penalty goals']/(euro['Penalty goals'] + euro['Penalties not scored'])
penalty_scoring_chance.index = euro['Team']
print(penalty_scoring_chance)
```

```python
goal_after_cross_chance = euro['Headed goals']/euro['Crosses']
goal_after_cross_chance.index = euro['Team']
print(goal_after_cross_chance.sort_values(ascending = False))
```

```python
chance_of_y_card = euro['Yellow Cards']/euro['Fouls Conceded']
chance_of_y_card.index = euro['Team']
print(chance_of_y_card.sort_values(ascending = False))
```

```python
touch_poss = euro['Goals']/euro['Touches'] #possibility of touch becoming resultative
touch_poss.index = euro['Team']
print(touch_poss.sort_values(ascending = False))
```
